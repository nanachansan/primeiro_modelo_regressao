```python
#para importar e manipular o df
import pandas as pd

#para criar e validar o modelo antes de passar o dataset de teste oficial
from sklearn.model_selection import train_test_split

#para fazer a regressão
from sklearn.ensemble import RandomForestRegressor

#para ajustar o modelo conforme o mae
from sklearn.metrics import mean_absolute_error

from sklearn.impute import SimpleImputer
```

```python
#colocar o path dentro de uma variável para chamar mais fácil
train_set = "C:/Users/IP/Documents/ACCELERADEV/train.csv"

test_set = "C:/Users/IP/Documents/ACCELERADEV/test.csv"

#ler o csv pelo pandas
dados_treino = pd.read_csv(train_set)
dados_teste = pd.read_csv(test_set)
```

```python
dados_treino.columns
```

```python
colunas_teste_pra_treino = ['NU_INSCRICAO', 'CO_UF_RESIDENCIA', 'SG_UF_RESIDENCIA', 'NU_IDADE',
       'TP_SEXO', 'TP_COR_RACA', 'TP_NACIONALIDADE', 'TP_ST_CONCLUSAO',
       'TP_ANO_CONCLUIU', 'TP_ESCOLA', 'TP_ENSINO', 'IN_TREINEIRO',
       'TP_DEPENDENCIA_ADM_ESC', 'IN_BAIXA_VISAO', 'IN_CEGUEIRA', 'IN_SURDEZ',
       'IN_DISLEXIA', 'IN_DISCALCULIA', 'IN_SABATISTA', 'IN_GESTANTE',
       'IN_IDOSO', 'TP_PRESENCA_CN', 'TP_PRESENCA_CH', 'TP_PRESENCA_LC',
       'CO_PROVA_CN', 'CO_PROVA_CH', 'CO_PROVA_LC', 'CO_PROVA_MT',
       'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'TP_LINGUA',
       'TP_STATUS_REDACAO', 'NU_NOTA_COMP1', 'NU_NOTA_COMP2', 'NU_NOTA_COMP3',
       'NU_NOTA_COMP4', 'NU_NOTA_COMP5', 'NU_NOTA_REDACAO', 'Q001', 'Q002',
       'Q006', 'Q024', 'Q025', 'Q026', 'Q027', 'Q047', 'NU_NOTA_MT']

dados_treino_ajustados = dados_treino.filter(colunas_teste_pra_treino)

dados_treino_ajustados.head()
```

```python
features = [ "NU_NOTA_CN", "NU_NOTA_CH", "NU_NOTA_LC", "NU_NOTA_COMP1", 
            "NU_NOTA_COMP2", "NU_NOTA_COMP3", "NU_NOTA_COMP4", "NU_NOTA_COMP5", "NU_NOTA_REDACAO", 
            "CO_UF_RESIDENCIA", "NU_NOTA_MT"]
meu_inputador = SimpleImputer()
features_para_ajustar = dados_treino_ajustados[features]

dados_treino_ajustados_ = pd.DataFrame(meu_inputador.fit_transform(features_para_ajustar))

dados_treino_ajustados_.columns = features_para_ajustar.columns

dados_treino_ajustados_
```

```python
features = [ "NU_NOTA_CN", "NU_NOTA_CH", "NU_NOTA_LC", "NU_NOTA_COMP1", 
            "NU_NOTA_COMP2", "NU_NOTA_COMP3", "NU_NOTA_COMP4", "NU_NOTA_COMP5", "NU_NOTA_REDACAO", 
            "CO_UF_RESIDENCIA"]

#definir o X e o y para treinar o modelo

y = dados_treino_ajustados_.NU_NOTA_MT #target 
X = dados_treino_ajustados_[features] #features

#hora de usar o train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=1)






```

```python
regressor_rf = RandomForestRegressor(random_state=0)

regressor_rf.fit(X_train, y_train)

preds = regressor_rf.predict(X_val)

mae = mean_absolute_error(y_val, preds)

```

```python
features_para_ajustar = dados_teste[features]

dados_teste2 = pd.DataFrame(meu_inputador.fit_transform(features_para_ajustar))

dados_teste2.columns = features_para_ajustar.columns


dados_teste.merge(dados_teste2)
dados_teste.isnull().sum()
```

```python
preds_final = regressor_rf.predict(dados_teste2)


output = pd.DataFrame({'NU_INSCRICAO': dados_teste.NU_INSCRICAO,
                       'NU_NOTA_MT': preds_final})
output.to_csv('answer.csv', index=False)

output.describe()
```
